```r
L=10
n=3
p=0.05
k=4
av=7.651
tmp=dbinom(x=0:(L-n), size=(L-n), prob=((av-n)/(L-n)))
total_list=c()
for(x in 0:(L-n)){
    proportion=tmp[x+1]/sum(tmp)
    sequenced_error=proportion*(x+n)*(p)
    non_sequenced_error=proportion*(L-x-n)*(1-0.25)
    print(paste((x+n), (L-x-n), proportion, sequenced_error, non_sequenced_error))
    summation=sequenced_error+non_sequenced_error
    total_list=c(total_list,summation)
}
print(sum(total_list))
```
